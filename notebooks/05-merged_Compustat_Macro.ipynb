```python
import pandas as pd
from IPython.display import display

# ============================================================
# 0. Load cleaned firm-level and macro datasets
# ============================================================

kpi_clean_path = "../data/processed/compustat_kpis_clean.csv"
kpi_scaled_path = "../data/processed/compustat_kpis_clean_scaled.csv"

macro_clean_path = "../data/processed/macro_clean.csv"
macro_scaled_path = "../data/processed/macro_clean_scaled.csv"

df_firm = pd.read_csv(kpi_clean_path)
df_firm_scaled = pd.read_csv(kpi_scaled_path)

df_macro = pd.read_csv(macro_clean_path)
df_macro_scaled = pd.read_csv(macro_scaled_path)

print("Firm-level clean shape:", df_firm.shape)
print("Macro clean shape:", df_macro.shape)

# ============================================================
# 1. Prepare merge keys
# ============================================================

df_firm["fyear"] = pd.to_numeric(df_firm["fyear"], errors="coerce").astype("Int64")
df_firm_scaled["fyear"] = pd.to_numeric(df_firm_scaled["fyear"], errors="coerce").astype("Int64")

df_macro["year"] = pd.to_numeric(df_macro["year"], errors="coerce").astype("Int64")
df_macro_scaled["year"] = pd.to_numeric(df_macro_scaled["year"], errors="coerce").astype("Int64")

print("\nFirm fyear range:", df_firm["fyear"].min(), "→", df_firm["fyear"].max())
print("Macro year range:", df_macro["year"].min(), "→", df_macro["year"].max())

# ============================================================
# 2. Identify macro columns
# ============================================================

macro_id_cols = ["year"]
macro_var_cols = [c for c in df_macro.columns if c not in macro_id_cols]

macro_scaled_var_cols = [c for c in df_macro_scaled.columns if c not in ["year"]]

# ============================================================
# 3. Merge UN-SCALED: firm_clean + macro_clean
# ============================================================

df_merged = df_firm.merge(
    df_macro,
    left_on="fyear",
    right_on="year",
    how="left",
    suffixes=("", "_macro")
)

if "year" in df_merged.columns:
    df_merged = df_merged.drop(columns=["year"])

print("\nMerged (unscaled) dataset shape:", df_merged.shape)
display(df_merged.head())

# ============================================================
# 4. Merge SCALED: firm_scaled + macro_scaled
# ============================================================

df_merged_scaled = df_firm_scaled.merge(
    df_macro_scaled,
    left_on="fyear",
    right_on="year",
    how="left",
    suffixes=("", "_macro")
)

if "year" in df_merged_scaled.columns:
    df_merged_scaled = df_merged_scaled.drop(columns=["year"])

print("\nMerged (scaled) dataset shape:", df_merged_scaled.shape)
display(df_merged_scaled.head())

# ============================================================
# 5. Save merged datasets (NOW INTO /data/final/)
# ============================================================

output_merged = "../data/final/firm_macro_merged.csv"
output_merged_scaled = "../data/final/firm_macro_merged_scaled.csv"

df_merged.to_csv(output_merged, index=False)
df_merged_scaled.to_csv(output_merged_scaled, index=False)

print("\nSaved merged unscaled dataset to:", output_merged)
print("Saved merged scaled dataset to:", output_merged_scaled)

# ============================================================
# Summary
# ============================================================

print("\nSummary:")
print("  Firm-level clean:", df_firm.shape)
print("  Macro clean:", df_macro.shape)
print("  Merged unscaled:", df_merged.shape)
print("  Merged scaled:", df_merged_scaled.shape)
```
